```python
import numpy as np
import matplotlib.pyplot as plt
def f(x): return x+1

def g(x): return x**2

def h(x): return f(g(x))

x = np.array(range(-10,10))

# 这里我们使用了Python的list comprehension来计算y
y = np.array([h(i) for i in x])

# 'bo' 将表示我们会使用蓝色的圆圈绘制点图，而非默认的线图
plt.plot(x, y, 'bo')
```
