```python
def f(x):
    return x**3 - 5*x**2 + 9
```

```python
f(3)
```

```python
import numpy as np
x = np.linspace(-5, 5, num = 100)
y = f(x)
import matplotlib.pyplot as plt
plt.plot(x,y)
```

```python
def exp2(x):
    sum = 0
    for k in range(100):
        sum += float(x**k)/np.math.factorial(k)
    return sum
```

```python
exp2(1),exp2(2)
```
